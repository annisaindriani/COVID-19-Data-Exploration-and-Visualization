### 1. Jumlah kasus terkonfirmasi COVID-19 berdasarkan negara

**Informasi diatas masih sangat general sehingga anda perlu mendapatkan informasi yang lebih spesifik, salah satunya mendapatkan informasi jumlah kasus terkonfirmasi COVID-19 berdasarkan kriteria tertentu. Anda perlu menampilkan dataframe`confirmed_country`yang berisi jumlah kasus terkonfirmasi COVID-19 di setiap negara berdasarkan deret waktu (time series) yang terindeks berdasarkan waktu(date/time) bukan berdasarkan `Country/Region`**.

```python
# Groupby berdasarkan Country/Region, kemudian transpose sehingga data terindeks berdasarkan waktu
raw_confirmed_country = raw_data_confirmed.groupby('Country/Region').sum().T

# Menghapus kolom Lat dan Long karena tidak diperlukan
confirmed_country = raw_confirmed_country.drop(['Lat', 'Long'])

# Mengubah waktu menjadi tipe data DateTime
confirmed_country_time = pd.to_datetime(confirmed_country.index)
datetime_index = pd.DatetimeIndex(confirmed_country_time.values) 
confirmed_country = confirmed_country.set_index(datetime_index)

# Menghapus nama kolom
confirmed_country.columns.name = ""
confirmed_country
```

##### Tuliskan analisis anda berdasarkan hasil diatas:  
Di banyak negara, pada bulan Januari jumlah kasus terkonfirmasi COVID-19 masih sedikit. Namun,
bisa dilihat seiring perkembangannya waktu kasus terkonfirmasi COVID-19 meningkat di beberapa negara.

### 2. Visualisasi kasus terkonfirmasi COVID-19 berdasarkan negara

##### Tuliskan analisis anda berdasarkan hasil visualisasi diatas:

### 4. Penyelarasan kurva pertumbuhan seluruh negara dengan jumlah kasus kematian COVID-19 â‰¥ 25

### 5. Visualisasi jumlah pasien yang sembuh (recovered people) 
    
**Anda perlu mendapatkan informasi mengenai total pasien COVID-19 yang dinyatakan sembuh seperti pada kasus terkonfirmasi dan kasus kematian tapi, kali ini lakukan hal tersebut dengan membuat `user-defined functions` yang diperlukan bukan mengetik ulang barisan code yang sama seperti di nomor sebelumnya.**

```python
def statistics_data(data):
    return data.describe()

def inform_data(data):
    return data.info()

def manipulation(data):
    recovered = data.groupby('Country/Region').sum().T
    recovered = recovered.drop(['Lat', 'Long'])
    
    recovered_time = pd.to_datetime(recovered.index)
    datetime_index = pd.DatetimeIndex(recovered_time.values) 
    recovered = recovered.set_index(datetime_index)

    recovered.columns.name = ""
    return recovered
```

**Periksa fungsi-fungsi yang sudah anda buat**

```python
# Function untuk menampilkan rangkuman statistik dari jumlah pasien COVID-19 yang sembuh
statistics_data(raw_data_recovered)
```

```python
# Function untuk menampilkan informasi mengenai data dari jumlah pasien COVID-19 yang sembuh
inform_data(raw_data_recovered)
```

```python
# Function untuk menampilkan data dari jumlah pasien kasus COVID-19 yang telah sembuh yang telah diubah sesuai indeks deret waktu dan sesuai format date time
manipulation(raw_data_recovered)
```

**Selanjutnya, tampilkan data total pasien COVID-19 yang sembuh dalam bentuk grafik yang sesuai. Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
# Function untuk menampilkan visualisasi jumlah pasien sembuh COVID-19
def visualization(data, title, xlabel, ylabel):
    plt = data.drop(['Lat', 'Long', 'Province/State', 'Country/Region'], axis = 1)
    return plt.sum().plot(title = title, xlabel = xlabel, ylabel = ylabel, figsize = (10,6))

visualization(raw_data_recovered, 'Jumlah Pasien Sembuh COVID-19', 'Date', 'Total')
```

##### Tuliskan analisis anda berdasarkan hasil visualisasi di atas:
Jumlah pasien yang sembuh dari COVID-19 terus mengalami peningkatan. Penurunan yang terlihat dalam line chart tersebut berbanding lurus dengan keadaan pada saat itu dimana kasus terkonfirmasi COVID-19 juga mengalami penurunan.

**Lakukan visualisasi yang sama pada beberapa negara berikut (France, Spain, China, US, Italy, and Australia). Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
# Function untuk menampilkan visualisasi pasien yang sembuh dari COVID-19 di beberapa negara
def visualization_country(data, country, title, xlabel, ylabel):
    data[country].plot(title = title, xlabel = xlabel, ylabel = ylabel, figsize = (10,6))
                                                                                      
visualization_country(manipulation(raw_data_recovered), ['France','Spain','China','US','Italy','Australia'], 'Jumlah Pasien Sembuh COVID-19', 'Date', 'Total')
```

Hampir di seluruh negara yang disebutkan, jumlah pasien sembuh COVID-19 mengalami kenaikan. Penurunan tajam pada bulan Januari 2021 di negara Amerika Serikat dan Italy menunjukkan bahwa jumlah pasien sembuh COVID-19 berbanding lurus dengan jumlah terkonfirmasi COVID-19 pada saat itu yang telah menurun.
